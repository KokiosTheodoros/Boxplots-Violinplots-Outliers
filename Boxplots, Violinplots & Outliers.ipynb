```python
# Day 5 - Boxplots, Violinplots & Outliers

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Φορτώνουμε dataset (Iris)
df = sns.load_dataset("iris")

# 1. Boxplot: Κατανομή petal_length ανά species
plt.figure(figsize=(8,5))
sns.boxplot(x="species", y="petal_length", data=df, palette="Set2")
plt.title("Boxplot: Petal Length per Species")
plt.show()

# 2. Violinplot: Petal length ανά species
plt.figure(figsize=(8,5))
sns.violinplot(x="species", y="petal_length", data=df, palette="Set2")
plt.title("Violinplot: Petal Length per Species")
plt.show()

# 3. Boxplot + Swarmplot: Στατιστικά + δεδομένα
plt.figure(figsize=(8,5))
sns.boxplot(x="species", y="petal_length", data=df, palette="Set2")
sns.swarmplot(x="species", y="petal_length", data=df, color="black", size=3)
plt.title("Boxplot + Swarmplot: Petal Length per Species")
plt.show()

# 4. Εντοπισμός Outliers (IQR)
Q1 = df['petal_length'].quantile(0.25)
Q3 = df['petal_length'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['petal_length'] < Q1 - 1.5*IQR) | (df['petal_length'] > Q3 + 1.5*IQR)]
print(" Outliers found in petal_length:")
print(outliers)
```
